# Depuración Técnica: Tratamiento de Valores Faltantes (NaN)

En esta fase del flujo de trabajo de Minería de Datos, se procede a la creación del objeto **icfes_sin_nan**. Este proceso consiste en filtrar la base de datos original para conservar únicamente los registros que contienen información completa en las seis dimensiones críticas de este estudio:

- Variable respuesta: Nivel Socioeconómico (ESTU_INSE_INDIVIDUAL)

  - Lectura Crítica

  - Inglés

# Justificación Técnica

La eliminación de los valores faltantes (NaN) no es un proceso arbitrario, sino una necesidad estadística fundamentada en los siguientes pilares:

## Requerimientos de los Algoritmos
Los modelos que se implementarán posteriormente, como la **Regresión Lineal Múltiple** y la **Red Neuronal de Perceptrón Multicapa (MLPRegressor)**, se basan en operaciones matriciales que no pueden procesar valores nulos.

La presencia de un solo NaN en una fila invalidaría el cálculo de los pesos y la convergencia del gradiente en la red neuronal.

## Validez de la Inferencia
Para que los coeficientes de correlación de **Pearson**, **Kendall** y **Spearman** sean comparables y válidos, deben calcularse sobre el mismo conjunto de observaciones.

Limpiar los datos asegura que la estructura de dependencia capturada no esté sesgada por la ausencia sistemática de información en ciertos estratos.

## Preparación para la Imputación
Este dataset depurado (**icfes_sin_nan**) servirá como la base de conocimiento para entrenar el modelo que finalmente imputará (predecirá) los valores faltantes en el grupo **icfes_connan**.

Esto garantiza que dichas predicciones tengan un respaldo estadístico sólido.

```python
# Siguiente paso: Eliminar filas donde falten datos clave
icfes_sin_nan = datos_icfes[(~datos_icfes['ESTU_INSE_INDIVIDUAL'].isna()) &
                            (~datos_icfes['PUNT_LECTURA_CRITICA'].isna()) &
                            (~datos_icfes['PUNT_MATEMATICAS'].isna()) &
                            (~datos_icfes['PUNT_C_NATURALES'].isna()) &
                            (~datos_icfes['PUNT_SOCIALES_CIUDADANAS'].isna()) &
                            (~datos_icfes['PUNT_INGLES'].isna())]
```

La visualización en 3D confirma que Lectura y Matemáticas son predictores adecuados. Por lo tanto, se procede a la fase de entrenamiento dividiendo los datos (80/20) para implementar los modelos de aprendizaje de máquina (Regresión y Red Neuronal) solicitados en la guía.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Definimos variable respuesta y explicativas
y = icfes_sin_nan['ESTU_INSE_INDIVIDUAL']
explicativas = ['PUNT_LECTURA_CRITICA', 'PUNT_MATEMATICAS', 'PUNT_C_NATURALES',
                'PUNT_SOCIALES_CIUDADANAS', 'PUNT_INGLES']

# Función para calcular R^2 omitiendo una variable
for var_omitida in explicativas:
    vars_actuales = [v for v in explicativas if v != var_omitida]
    X_temp = icfes_sin_nan[vars_actuales]
    modelo_temp = LinearRegression().fit(X_temp, y)
    print(f"R^2 omitiendo {var_omitida}: {modelo_temp.score(X_temp, y):.4f}")

# R^2 con todas las variables para comparar
X_todas = icfes_sin_nan[explicativas]
modelo_todas = LinearRegression().fit(X_todas, y)
print(f"R^2 con TODAS las variables: {modelo_todas.score(X_todas, y):.4f}")
```

# Análisis de los Resultados

Al observar la salida del código, se extraen conclusiones críticas de profundidad estadística:

## Variables Redundantes (Ciencias y Sociales)
Se observa que al omitir **Ciencias Naturales** o **Sociales y Ciudadanas**, el R² se mantiene prácticamente igual al modelo completo (**0.3060**).  

Estadísticamente, esto sugiere que estas áreas presentan **multicolinealidad alta** con las demás; es decir, no añaden capacidad predictiva nueva al modelo lineal.

## Ajuste Global
El modelo con todas las variables logra explicar un **30.60%** de la variabilidad del nivel socioeconómico.  

Para un fenómeno social tan complejo, este valor es significativo, aunque justifica la necesidad de usar una **Red Neuronal** más adelante para intentar capturar relaciones no lineales que puedan aumentar este ajuste.

# Fase de Preparación: División del Dataset (Train/Test Split)

Una vez seleccionadas las variables con mayor capacidad explicativa, el flujo de trabajo de la minería de datos exige dividir la información en dos subconjuntos independientes.

Esta separación es fundamental para entrenar el **Modelo de Regresión Lineal Múltiple** y la **Red Neuronal (MLPRegressor)**, permitiendo evaluar su desempeño real con datos que los algoritmos no han visto previamente.

# Descripción Línea a Línea del Código

## `from sklearn.model_selection import train_test_split`
Se importa la función estándar de la librería **scikit-learn** para realizar particiones aleatorias de datos.

## `X = icfes_sin_nan[explicativas]`
Se define la matriz de características (variables independientes) que contiene los puntajes de las 5 áreas académicas.

## `y = icfes_sin_nan['ESTU_INSE_INDIVIDUAL']`
Se define el vector de etiquetas (variable respuesta) correspondiente al Nivel Socioeconómico.

## `test_size=0.20`
Se especifica que el 20% de los datos se reservará para la validación, dejando el 80% restante para el aprendizaje.

## `random_state=42`
Se fija una "semilla" de aleatoriedad. Esto es una buena práctica estadística que garantiza que los resultados sean reproducibles; es decir, que cada vez que ejecutes el código, los grupos de entrenamiento y prueba contengan exactamente los mismos registros.

```python
from sklearn.model_selection import train_test_split

X = icfes_sin_nan[explicativas]
y = icfes_sin_nan['ESTU_INSE_INDIVIDUAL']

# División 80% entrenamiento, 20% validación
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

print(f"Tamaño entrenamiento: {X_train.shape} registros")
print(f"Tamaño validación: {X_test.shape} registros")
```

# Análisis de los Resultados

# Implementación de Modelos Predictivos: Regresión vs. Red Neuronal

En esta etapa se entrenan dos algoritmos de distinta naturaleza: un modelo paramétrico tradicional (**Regresión Lineal Múltiple**) y un modelo no paramétrico de aprendizaje profundo (**Red Neuronal de Perceptrón Multicapa**).

El objetivo es determinar cuál de las dos arquitecturas logra capturar con mayor precisión la variabilidad de las condiciones socioeconómicas de los estudiantes evaluados por el ICFES.

---

# Descripción Línea a Línea del Código

## `from sklearn.neural_network import MLPRegressor`
Se importa el algoritmo de redes neuronales diseñado para predecir variables continuas (regresión).

## `from sklearn.metrics import r2_score`
Se carga la métrica del Coeficiente de Determinación, herramienta estándar para medir la calidad del ajuste del modelo.

## `modelo_lin.fit(X_train, y_train)`
Fase de aprendizaje del modelo lineal, donde se calculan los coeficientes (β) que minimizan el error cuadrático medio sobre los 9,482 registros de entrenamiento.

## `MLPRegressor(hidden_layer_sizes=(100,100,100,100, 100, 100), ...)`
Se define la arquitectura de la red neuronal siguiendo los parámetros establecidos: 6 capas ocultas con 100 neuronas cada una, con el objetivo de maximizar la capacidad de detección de patrones complejos.

## `max_iter=500, activation='relu', solver='adam'`
Se configuran los hiperparámetros de optimización:

- `relu`: permite manejar la no linealidad del problema  
- `adam`: asegura una convergencia eficiente del gradiente  
- `max_iter=500`: limita las iteraciones para el entrenamiento del modelo  

## `r2_score(y_test, y_pred)`
Se calcula el ajuste final comparando las predicciones del modelo con los valores reales del 20% de datos reservados (validación).

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

# 1. Modelo de Regresión Lineal Múltiple
modelo_lin = LinearRegression()
modelo_lin.fit(X_train, y_train)
y_pred_lin = modelo_lin.predict(X_test)

# 2. Modelo de Red Neuronal (Parámetros de la guía [1])
red = MLPRegressor(hidden_layer_sizes=(100,100,100,100, 100, 100),
                   max_iter=500,
                   activation='relu',
                   solver='adam',
                   learning_rate='invscaling',
                   random_state=42)
red.fit(X_train, y_train)
y_pred_red = red.predict(X_test)

print(f"R^2 Regresión Lineal: {r2_score(y_test, y_pred_lin):.4f}")
print(f"R^2 Red Neuronal: {r2_score(y_test, y_pred_red):.4f}")
```

# Análisis de los Resultados

Tras ejecutar los modelos, se obtuvieron los siguientes coeficientes de determinación:

- **R² Regresión Lineal:** 0.3198 (31.98%)  
- **R² Red Neuronal:** 0.2976 (29.76%)  

---

## Interpretación Técnica

### Capacidad Explicativa
Ambos modelos logran explicar aproximadamente el **30-32% de la varianza del nivel socioeconómico**. Esto confirma estadísticamente que, si bien existe una relación real entre el entorno del hogar y el éxito académico, esta asociación es moderada y existen factores externos no capturados en las pruebas que influyen en el NSE.

---

### Comparación de Modelos
Sorprendentemente, la **Regresión Lineal Múltiple** superó ligeramente a la **Red Neuronal** en este dataset (0.3198 vs 0.2976).

Esto sugiere que la relación entre los puntajes de las áreas básicas y el nivel socioeconómico es mayoritariamente lineal, y que la alta complejidad de la red neuronal (600 neuronas totales) podría estar generando un ligero subajuste (underfitting) o que la arquitectura es demasiado densa para la cantidad de variables explicativas (5) utilizadas.

---

### Justificación de Elección
Aunque el R² lineal es mayor, para propósitos de imputación de datos faltantes, se seleccionará el modelo con el coeficiente más alto (**Regresión Lineal**) para garantizar que los valores predichos en el grupo **icfes_connan** tengan el menor error posible.
